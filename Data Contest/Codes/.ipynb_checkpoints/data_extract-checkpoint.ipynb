```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("ratings.csv")
remarks = pd.read_csv("remarks.csv")
remarks_supp_opp = pd.read_csv("remarks_supp_opp.csv")
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
ratings.head()
```

```python
ratings["Date"] = pd.to_datetime(ratings["Date"])
ratings.insert(0, 'my_id', ratings['emp'].apply(str) + ratings['comp'])
```

```python

grouped_ratings = ratings.groupby('my_id')
df = grouped_ratings.max()['Date']
df = df.reset_index()
df.columns = ['my_id','last_rating_date']
df_2 = grouped_ratings.min()['Date']
df_2 = df_2.reset_index()
df_2.columns = ['my_id','first_rating_date']
grouped_dates = df.join(df_2.set_index('my_id'),on='my_id')
grouped_dates.head()
```

```python
min_date = ratings['Date'].min()
max_date = ratings['Date'].max()
delta_time = float((max_date - min_date).days)
ratings['Weights'] = ratings['Date'].apply(lambda x: float((x-min_date).days)/delta_time)
# ratings
```

```python
ratings['prod'] = ratings['rating']*ratings['Weights']

grouped_no_ratings = ratings.groupby('my_id').size()
grouped_no_ratings = grouped_no_ratings.reset_index()
grouped_no_ratings.columns = ['my_id','no_ratings']

grouped_sum_weights = ratings.groupby('my_id').sum()['Weights']
grouped_sum_weights = grouped_sum_weights.reset_index()
grouped_sum_weights.columns = ['my_id','sum_weights']

grouped_sum_prod = ratings.groupby('my_id').sum()['prod']
grouped_sum_prod = grouped_sum_prod.reset_index()
grouped_sum_prod.columns = ['my_id','sum_prod']

grouped_sum = grouped_sum_weights.join(grouped_sum_prod.set_index('my_id'),on='my_id')
grouped_sum['weighted_rating'] = grouped_sum['sum_prod']/grouped_sum['sum_weights']

grouped_weighted_rating = grouped_sum.drop(['sum_weights','sum_prod'],axis = 1)
grouped_weighted_rating = grouped_dates.join(grouped_weighted_rating.set_index('my_id'),on='my_id')
grouped_weighted_rating = grouped_weighted_rating.join(grouped_no_ratings.set_index('my_id'),on='my_id')
grouped_weighted_rating
```

```python
# Dropping Negative ID's
remarks_supp_opp = remarks_supp_opp.drop(remarks_supp_opp[remarks_supp_opp.emp<0].index)
```

```python
# Inserting my_id
remarks_supp_opp.insert(0, 'my_id', remarks_supp_opp['emp'].apply(str) + remarks_supp_opp['comp'])
train.insert(0, 'my_id', train['emp'].apply(str) + train['comp'])
```

```python
# Transforming remarks.csv
remarks = pd.read_csv("remarks.csv")

# Dropping Negative ID's and duplicates
remarks = remarks.drop(remarks[remarks.emp<0].index)
remarks = remarks.drop_duplicates()

# Inserting 'my_id'
remarks.insert(0, 'my_id', remarks['emp'].apply(str) + remarks['comp'])

# txt to length
remarks['txt'] = remarks['txt'].apply(str)
remarks['txt'] = remarks['txt'].apply(lambda x: len(x))

remarks['remarkDate'] = pd.to_datetime(remarks['remarkDate'])
```

```python
# Transforming remarks_supp_opp.csv
remarks_supp_opp = pd.read_csv("remarks_supp_opp.csv")

# Dropping Negative ID's
remarks_supp_opp = remarks_supp_opp.drop(remarks_supp_opp[remarks_supp_opp.emp<0].index)

# Inserting 'my_id'
remarks_supp_opp.insert(0, 'my_id', remarks_supp_opp['emp'].apply(str) + remarks_supp_opp['comp'])

# True/False to 1/0
remarks_supp_opp['support'] = remarks_supp_opp['support'].apply(lambda x: 1 if (x == True) else 0)
remarks_supp_opp['oppose'] = remarks_supp_opp['oppose'].apply(lambda x: 1 if (x == True) else 0)
```

```python
short_train = train.drop(['id','emp','comp','lastratingdate'] ,axis=1)
recd_remarks = pd.merge(remarks_supp_opp,short_train,how='inner',on='my_id')

recd_remarks = recd_remarks.drop_duplicates()
recd_remarks = recd_remarks.sort_values('remarkId',axis=0)
pivot_recd_remarks = pd.pivot_table(recd_remarks, values=['support','oppose'], index='remarkId',
                    columns=['left'], aggfunc=np.sum,fill_value = 0)
```

```python
recd_remarks_supp_opp = pivot_recd_remarks.reset_index()
recd_remarks_supp_opp.columns = ['remarkId','recd_opp_0','recd_opp_1','recd_supp_0','recd_supp_1']
```

```python
df = pd.merge(recd_remarks_supp_opp,remarks,how='right',on='remarkId')
df['recd_opp_0'] = df['recd_opp_0'].apply(lambda x: 0 if (pd.isnull(x)) else x)
df['recd_opp_1'] = df['recd_opp_1'].apply(lambda x: 0 if (pd.isnull(x)) else x)
df['recd_supp_0'] = df['recd_supp_0'].apply(lambda x: 0 if (pd.isnull(x)) else x)
df['recd_supp_1'] = df['recd_supp_1'].apply(lambda x: 0 if (pd.isnull(x)) else x)

min_date = df['remarkDate'].min()
max_date = df['remarkDate'].max()
delta_time = float((max_date - min_date).days)
df['Weights'] = (df['remarkDate'].apply(lambda x: float((x-min_date).days)/delta_time))*df['txt']
df['recd_opp_0'] = df['recd_opp_0']*df['Weights']
df['recd_opp_1'] = df['recd_opp_1']*df['Weights']
df['recd_supp_0'] = df['recd_supp_0']*df['Weights']
df['recd_supp_1'] = df['recd_supp_1']*df['Weights']

df_weighted = df.groupby('my_id')['recd_opp_0', 'recd_opp_1','recd_supp_0','recd_supp_1','Weights'].sum().reset_index()
df_weighted['recd_opp_0'] = df_weighted['recd_opp_0']/df_weighted['Weights']
df_weighted['recd_opp_1'] = df_weighted['recd_opp_1']/df['Weights']
df_weighted['recd_supp_0'] = df_weighted['recd_supp_0']/df_weighted['Weights']
df_weighted['recd_supp_1'] = df_weighted['recd_supp_1']/df_weighted['Weights']

df_weighted['last_remark_date'] =df.groupby('my_id').max()['remarkDate'].reset_index()['remarkDate'].reset_index().drop(['index'],axis=1)
df_weighted['first_remark_date'] =df.groupby('my_id').min()['remarkDate'].reset_index()['remarkDate'].reset_index().drop(['index'],axis=1)
```

```python
recd_grouped_weighted_remark = df_weighted.drop(['Weights'],axis=1)
recd_grouped_weighted_remark
```

```python
remarks_supp_opp.sort_values('my_id')
```

```python
df_2 = pd.merge(remarks,short_train,how="inner",on='my_id')
df_2 = df_2.drop(['my_id','emp','comp'],axis=1)
df_3 = pd.merge(remarks_supp_opp,df_2,how="inner",on='remarkId')
df_3.sort_values('my_id')
pivot_given_remarks = pd.pivot_table(df_3, values=['support','oppose'], index='my_id',
                    columns=['left'], aggfunc=np.sum,fill_value = 0)
given_remarks_supp_opp = pivot_given_remarks.reset_index()

given_remarks_supp_opp.columns = ['my_id','giv_opp_0','giv_opp_1','giv_supp_0','giv_supp_1']
given_remarks_supp_opp
```

```python
train_id = short_train.drop(['left'],axis=1).drop_duplicates()
ratings_on_train = pd.merge(grouped_weighted_rating,train_id,how='right',on='my_id')
recd_remarks_on_train = pd.merge(recd_grouped_weighted_remark,train_id,how='right',on='my_id')
given_remarks_on_train = pd.merge(given_remarks_supp_opp,train_id,how='right',on='my_id')
remarks_on_train = pd.merge(recd_remarks_on_train,given_remarks_on_train,how='inner',on='my_id')
```

```python
features_on_train = pd.merge(ratings_on_train,remarks_on_train,how='inner',on='my_id')
features_on_train = pd.merge(features_on_train,short_train.drop_duplicates(),how='inner',on='my_id')
```

```python
test = pd.read_csv("test.csv")
test.insert(0, 'my_id', test['emp'].apply(str) + test['comp'])
test.drop_duplicates()
```

```python
test = test['my_id'].reset_index().drop(['index'],axis =1)

ratings_on_test = pd.merge(grouped_weighted_rating,test,how='right',on='my_id')
recd_remarks_on_test = pd.merge(recd_grouped_weighted_remark,test,how='right',on='my_id')
given_remarks_on_test = pd.merge(given_remarks_supp_opp,test,how='right',on='my_id')
remarks_on_test = pd.merge(recd_remarks_on_test,given_remarks_on_test,how='right',on='my_id')

features_on_test = pd.merge(ratings_on_test,remarks_on_test,how='inner',on='my_id')
features_on_test = pd.merge(features_on_test,test.drop_duplicates(),how='inner',on='my_id')
features_on_test
```
